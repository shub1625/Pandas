```python
import pandas as pd
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent') # To read the data from csv
```

```python
filt = df['Country'] == 'India'
india_df = df.loc[filt]
india_df.head()
```

```python
# write DF to csv
india_df.to_csv('data/modified_data.csv')
```

```python
# To write into tab seprated file (tsv)
india_df.to_csv('data/modified.tsv', sep='\t')
```

```python
# To write to xls format , do pip install xlwt openpyxl xlrd first
india_df.to_excel('data/modified.xlsx')
```

```python
# To write to json format , default will be dict like , but if we want list like records we can pass orient argument
india_df.to_json('data/modified.json', orient='records' ,lines=True)
```

```python
# To connect to postgres DB
# First install SQLAlchemy (it is an ORM for python)
# And for postgres pip install psycopg2-binary
# Now run below imports
from sqlalchemy import create_engine
import psycopg2
```

```python
# create_engine will be used to connect to DB
# postgresql://dbname:dbpass@host_server:port_number/DBNAME
engine = create_engine('postgresql://postgres:dbpass@localhost:5432/pandas')
```

```python
# Write data to DB
india_df.to_sql('sample_table', engine , if_exists='replace')
```

```python
# read data
sql_df = pd.read_sql('sample_table',engine, index_col='Respondent')
```
